#Basic Tools of matplotlib

```python
housing = pd.DataFrame({
    'rooms' : [1,1,2,2,2,3,3,3,3],
    'price' : [100,105,205,210,300,310,390,400,410]
})
```

```python
housing
```

```python
plt.scatter(housing['rooms'],housing['price']) # or we can also use - plt.scatter(x = housing['rooms'],y = housing['price'])
```

#Style Calls
To explain the graph to a non coder or client

```python
plt.plot(x,y)
plt.title('Title')
plt.xlabel('Here is x column')
plt.ylabel('Here is y column')
```

```python
plt.scatter(housing['rooms'],housing['price'])
plt.title('Hotel')
plt.xlabel('Rooms')
plt.ylabel('Price')
```

To get a part of the graph/Zoomed perspective, use xlim or ylim

```python
plt.plot(x,y)

#Axis and tricks
plt.xlim(0,2)
plt.ylim(100,200)


plt.title('Title')
plt.xlabel('Here is x column')
plt.ylabel('Here is y column')
```

Color and styling

```python
plt.plot(x,y, color = 'red', marker = '*', markersize = 15, linestyle = '--')
plt.title('Title')
plt.xlabel('Here is x column')
plt.ylabel('Here is y column')
```

#Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('/content/heart.csv')
```

```python
df.head()
```

#Distribution Plot

Used to plot the continuous data.

```python
sns.distplot(df['age'])
```

```python
sns.displot(df['age'])
```

```python
sns.displot(df['age'], bins = 30)
```

```python
sns.displot(df['age'], bins = 30, color = 'red')
```

#Countplot

Used to plot the categorical data.

```python
sns.countplot(data=df, x='sex')
```

```python
sns.countplot(data=df, x='cp')
```

```python
sns.countplot(data=df, x = 'cp', hue = 'sex' )
```

```python
sns.countplot(data=df, x = 'cp', hue = 'sex' , palette = 'terrain')
```

#BoxPlot
To visualize the data for both cont and catg data

```python
sns.boxplot(x = 'sex', y = 'age', data=df)
```

```python
sns.boxplot(x = 'target', y = 'age', hue = 'sex', data=df)
```

#Scatter Plot
To visualize the data for both cont data

```python
sns.scatterplot(data = df, x = 'chol', y = 'trestbps',hue = 'sex')
```
